```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,6))
plt.bar(df['Age'], df['Total Amount'])
plt.xlabel('Age')
plt.ylabel('Total Amount')
plt.title('bar plot between Age and Total amount')
```

```python
category_totals = df.groupby('Product Category')['Total Amount'].sum()
```

```python
plt.figure(figsize=(12,6))
# colors = sns.color_palette('pastel', len(category_totals))
plt.figure(figsize=(12, 6))
plt.pie(category_totals, labels=category_totals.index, autopct='%1.1f%%')
plt.title('pie plot between gender and Total amount')
```

```python
gender_spend = df.groupby('Gender')['Total Amount'].sum()
plt.bar(gender_spend.index, gender_spend.values)
plt.xlabel("Gender")
plt.ylabel("Total Amount")
```

```python
category_gender_totals = df.groupby(['Product Category', 'Gender'])['Total Amount'].sum().unstack()

category_gender_totals.plot(kind='bar', figsize=(6,4))
plt.title('Total Amount Spent by Gender and Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Amount')
plt.legend(title='Gender')
plt.xticks(rotation=0)
```

# Product analysis

```python
avg_price = {}
for i in df['Product Category'].unique():
    cate_df = df[df['Product Category'] == i]
    avg_price[i] = cate_df['Price per Unit'].mean()
    
avg_price_list = pd.DataFrame(list(avg_price.items()), columns=['Category', 'Average price per unit'] )

plt.bar(avg_price_list['Category'], avg_price_list['Average price per unit'],color=['skyblue', 'lightgreen', 'salmon'])
plt.title('category wise avg price per unit')
plt.xlabel('Category')
plt.ylabel('avg price per unit')
print(avg_price_list)
```

```python
avg_quantity = {}
for i in df['Product Category'].unique():
    cate_df = df[df['Product Category'] == i]
    avg_quantity[i] = cate_df['Quantity'].sum()
    
# Create DataFrame from the dictionary
quantity_list = pd.DataFrame(list(avg_quantity.items()), columns=['Category', 'Total Quantity sold'])

# Print the DataFrame to verify its contents
print(quantity_list)

# Plot the bar chart using the correct DataFrame and column names
plt.figure(figsize=(10, 6))
plt.bar(quantity_list['Category'], quantity_list['Total Quantity sold'], color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Category Wise Total Quantity Sold')
plt.xlabel('Product Category')
plt.ylabel('Total Quantity Sold')

plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
revenue = {}
for i in df['Product Category'].unique():
    cate_df = df[df['Product Category'] == i]
    revenue[i] = cate_df['Total Amount'].sum()
    
revenue_list = pd.DataFrame(list(revenue.items()), columns=['Category', 'Total Revenue by Category'])

plt.bar(revenue_list['Category'], revenue_list['Total Revenue by Category'],color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Total Revenue by Category')
plt.xlabel('Product Category')
plt.ylabel('Total Revenue')
print(revenue_list)
```

# Sales trend over time

```python
df['month']= df.index.month
df
```

```python
df['month']= df.index.month

revenue_by_year = df.groupby('month')['Total Amount'].sum()

plt.figure(figsize=(6, 4))
plt.plot(revenue_by_year.index, revenue_by_year.values, marker='o', linestyle='-', color='b')
plt.title('Month vs. Total Revenue')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.xticks(revenue_by_year.index)  # Ensure all years are shown on the x-axis
plt.show()
```

```python
bins = [17, 24, 34, 44, 54, 64, 74, 80, float('inf')]  # Bins for age ranges
labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65-74', '75-80', '81+']  # Labels for age ranges

# Create age group column without modifying the original 'Age' column
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True)

# Display the DataFrame with Age Groups
print(df.head())
```

```python
age_group = df.groupby('Age Group')['Total Amount'].sum()

plt.figure(figsize=(12,6))
# colors = sns.color_palette('pastel', len(category_totals))
plt.figure(figsize=(12, 6))
plt.pie(age_group, labels=age_group.index, autopct='%1.1f%%')
plt.title('pie plot between Age group and Total amount')
```
