# 1. Customer Analysis

### a. Age Distribution

```python
age_total_amount = df.groupby('Age')['Total Amount'].sum().reset_index()

age_total_amount = age_total_amount.sort_values('Age')

plt.figure(figsize=(6,4))
plt.bar(age_total_amount['Age'], age_total_amount['Total Amount'], color='skyblue')
plt.xlabel('Age')
plt.ylabel('Total Amount')
plt.title('Total Amount Spent by Age')
plt.xticks(rotation=45)  
plt.grid(axis='y')  
plt.show()
```

### b. Spending on different category

```python
category_totals = df.groupby('Product Category')['Total Amount'].sum()
```

```python
plt.figure(figsize=(6,4))
# colors = sns.color_palette('pastel', len(category_totals))
plt.figure(figsize=(6,4))
plt.pie(category_totals, labels=category_totals.index, autopct='%1.1f%%')
plt.title('pie plot between gender and Total amount')
```

```python
category_totals = df.groupby('Product Category')['Quantity'].sum()
plt.figure(figsize=(6,4))
# colors = sns.color_palette('pastel', len(category_totals))
plt.figure(figsize=(6,4))
plt.pie(category_totals, labels=category_totals.index, autopct='%1.1f%%')
plt.title('pie plot between gender and Quantity')
```

### c. Spending by gender

```python
gender_spend = df.groupby('Gender')['Total Amount'].sum()
plt.figure(figsize=(4,4))
plt.bar(gender_spend.index, gender_spend.values)
plt.xlabel("Gender")
plt.ylabel("Total Amount")
```

```python
category_gender_totals = df.groupby(['Product Category', 'Gender'])['Total Amount'].sum().unstack()

category_gender_totals.plot(kind='bar', figsize=(6,4))
plt.title('Total Amount Spent by Gender and Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Amount')
plt.legend(title='Gender')
plt.xticks(rotation=0)
```

### d. Spending by age group

```python
bins = [17, 24, 34, 44, 54, 64, 74, 80, float('inf')]  # Bins for age ranges
labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65-74', '75-80', '81+']  # Labels for age ranges

# Create age group column without modifying the original 'Age' column
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True)

# Display the DataFrame with Age Groups
print(df.head())
```

```python
age_group = df.groupby('Age Group')['Total Amount'].sum()

plt.figure(figsize=(6,4))
plt.pie(age_group, labels=age_group.index, autopct='%1.1f%%')
plt.title('pie plot between Age group and Total amount')
```

#### By Customer analysis i found that:
- Spending on different categories are almost same but **electronics** has highest expenditure.
- The Quantity sold of **clothing** are highest.
- **Females** are spending money more then **males**.
- **Females** are spending money on **beauty** and **clothing** while **males** are more spending  more money on **electronics**.
- Maximum customers are of **25-54** age group


# 2. Product analysis

### a. Price Analysis

```python
avg_price = {}
for i in df['Product Category'].unique():
    cate_df = df[df['Product Category'] == i]
    avg_price[i] = cate_df['Price per Unit'].mean()
    
avg_price_list = pd.DataFrame(list(avg_price.items()), columns=['Category', 'Average price per unit'] )

plt.figure(figsize=(6,4))
plt.bar(avg_price_list['Category'], avg_price_list['Average price per unit'],color=['skyblue', 'lightgreen', 'salmon'])
plt.title('category wise avg price per unit')
plt.xlabel('Category')
plt.ylabel('avg price per unit')
plt.grid(axis='y')  
print(avg_price_list)
```

```python
avg_quantity = {}
for i in df['Product Category'].unique():
    cate_df = df[df['Product Category'] == i]
    avg_quantity[i] = cate_df['Quantity'].sum()
    
# Create DataFrame from the dictionary
quantity_list = pd.DataFrame(list(avg_quantity.items()), columns=['Category', 'Total Quantity sold'])

# Print the DataFrame to verify its contents
print(quantity_list)

# Plot the bar chart using the correct DataFrame and column names
plt.figure(figsize=(6,4))
plt.bar(quantity_list['Category'], quantity_list['Total Quantity sold'], color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Category Wise Total Quantity Sold')
plt.xlabel('Product Category')
plt.ylabel('Total Quantity Sold')

plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
revenue = {}
for i in df['Product Category'].unique():
    cate_df = df[df['Product Category'] == i]
    revenue[i] = cate_df['Total Amount'].sum()
    
revenue_list = pd.DataFrame(list(revenue.items()), columns=['Category', 'Total Revenue by Category'])

plt.bar(revenue_list['Category'], revenue_list['Total Revenue by Category'],color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Total Revenue by Category')
plt.xlabel('Product Category')
plt.ylabel('Total Revenue')
print(revenue_list)
```

### b. Monthly revenue trend by Category

```python
df['month'] = df.index.month

# Group by month and Product Category and calculate the total amount spent
revenue_categ = df.groupby(['month', 'Product Category'])['Total Amount'].sum().reset_index()

# Pivot the data to have separate columns for each product category
pivot_df = revenue_categ.pivot(index='month', columns='Product Category', values='Total Amount').fillna(0)

# Plot the data
plt.figure(figsize=(6,4))
plt.plot(pivot_df.index, pivot_df['Beauty'], label='Beauty', color='blue', marker='o', linestyle='-')
plt.plot(pivot_df.index, pivot_df['Clothing'], label='Clothing', color='red', marker='o', linestyle='-')
plt.plot(pivot_df.index, pivot_df['Electronics'], label='Electronics', color='green', marker='o', linestyle='-')
plt.title('Revenue by Category Over Time')
plt.xlabel('Month')
plt.ylabel('Total Amount Spent')
plt.legend(title='Product Category')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

```python
df['month'] = df.index.month

# Group by month and Product Category and calculate the total amount spent
revenue_categ = df.groupby(['month', 'Product Category'])['Quantity'].sum().reset_index()

# Pivot the data to have separate columns for each product category
pivot_df = revenue_categ.pivot(index='month', columns='Product Category', values='Quantity').fillna(0)

# Plot the data
plt.figure(figsize=(6,4))
plt.plot(pivot_df.index, pivot_df['Beauty'], label='Beauty', color='blue', marker='o', linestyle='-')
plt.plot(pivot_df.index, pivot_df['Clothing'], label='Clothing', color='red', marker='o', linestyle='-')
plt.plot(pivot_df.index, pivot_df['Electronics'], label='Electronics', color='green', marker='o', linestyle='-')
plt.title('Quantity by Category Over Month')
plt.xlabel('Month')
plt.ylabel('Quantity')
plt.legend(title='Product Category')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

#### By Product Analysis I found that:
- The **Average Price** of **Beauty** Category is comparatively high.
- The **Quantity sold** of **clothing** is high.
- **Total Revenue** from **clothing** and **electronics** are almost the **same**. 
- **Monthly Revenue** from the **clothing** and **Beauty** categories are nearly constant while **electronics** has no fixed trend sometime reaching the **highest** and **lowest** points. This may be due to festivals or events.
- Quantity sold of clothing are high but **Revenue** is less and it is opposite for **Electronics**.

# 3. Sales Performance

### a. Sales Trend

```python
df['month']= df.index.month
df
```

```python
df['month']= df.index.month

revenue_by_year = df.groupby('month')['Total Amount'].sum()

plt.figure(figsize=(6, 4))
plt.plot(revenue_by_year.index, revenue_by_year.values, marker='o', linestyle='-', color='b')
plt.title('Month vs. Total Revenue')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.xticks(revenue_by_year.index)  # Ensure all years are shown on the x-axis
plt.show()
```

### b. Gender wise Sales Trend

```python
df['month']= df.index.month
gender_spend_trend = df.groupby(['month', 'Gender'])['Total Amount'].sum().reset_index()


pivot_df = gender_spend_trend.pivot(index='month', columns='Gender', values='Total Amount').fillna(0)

# Plot the data
plt.figure(figsize=(6,4))
plt.plot(pivot_df.index, pivot_df['Male'], label='Male', color='blue', marker='o', linestyle='-')
plt.plot(pivot_df.index, pivot_df['Female'], label='Female', color='red', marker='o', linestyle='-')
plt.title('Total Amount Spent by Gender Over Time')
plt.xlabel('Month')
plt.ylabel('Total Amount Spent')
plt.legend(title='Gender')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

#### By Sales Performance I Found That:
- Revenue in **Feb**, **May**, and **Oct** is the **highest** because of high sales of **electronics** category products.
- **Females** are spending almost constantly throughout the year while **males** sometimes spend too much and sometimes the lowest amount.

# Recommendations
Based on the findings from the sales data analysis, the following recommendations are suggested:

**1.** **Targeted Marketing for Age Groups:**

    - Focus marketing efforts on the 25-54 age group, which constitutes the majority of customers. Tailored promotions and advertisements for this demographic can further boost sales.
**2.** **Promotional Campaigns for Electronics:**

    - Given the high expenditure on electronics, especially during certain months, enhance promotional activities during these peak periods. Special discounts or bundled offers can help capitalize on this trend.
**3.** **Boost Clothing Sales Revenue:**

    - While the quantity sold for clothing is high, the revenue is comparatively lower. Introduce higher-value clothing items or premium collections to increase the average transaction value in this category.
**4.** **Enhance Beauty Product Marketing:**

    - With beauty products showing comparatively high average prices, consider targeted advertising and loyalty programs for customers purchasing these items to maximize revenue from this segment.
